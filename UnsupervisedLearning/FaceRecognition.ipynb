### Importing Useful Libraries and Writing Helper Functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os,cv2


IMGDATA = "att_faces"
FOLDERS =[]
training_data=[]
for i in range(1,41):
    FOLDERS.append("s"+str(i))
```

```python
from PIL import Image
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = Image.open(os.path.join(folder,filename))
        img = np.array(img.convert('L')) #converting to Grayscale
        img = img / 255 
        print(img.shape)
        print(img)
        if img is not None:
            images.append(img)
    return images
```

## Data Loading and Preprocessing

```python
data=[]
for i in range(1,41):
    for j in load_images_from_folder(os.path.join(IMGDATA,"s"+str(i))):
        data.append(list((j,i)))
```

```python
imgdata = pd.DataFrame(data,columns=['Image','Label'])
```

```python
plt.imshow(imgdata['Image'][0])
plt.show()
```

```python
print(len(imgdata))
print(imgdata['Image'][0].shape)
```

<code>images are grayscale and 92 x 112 pixels in dimension</code>

<code>we need to reshape the matrix of pixel intensities for an image into a vector and create a matrix of these vectors for all the training images</code>

### Reshaping the matrix of pixel intensities for an image into a vector and create a matrix of these vectors for all the training images

```python
imagesCount=len(imgdata['Image'])
for i in range(imagesCount):
    img = imgdata['Image'][i]
    print(img.shape)
    imgdata['Image'][i] = img.reshape(img.shape[0]*img.shape[1])
```

```python
imgdata['Image'][0].shape
```

<code>We can see that each image of shape (112, 92) is transformed into a feature vector of 10304 dimensions</code>

```python
imgdata['Image'][1]
```

```python
train_img, test_img, train_lbl, test_lbl=([],[],[],[])
```

```python
muls=[10*i for i in range(1,41) ]
for i in range(0,400):
    if(i>0  and (i+1 in  muls)):
        test_img.append(imgdata['Image'][i])
        test_lbl.append(imgdata['Label'][i])
    else:
        train_img.append(imgdata['Image'][i])
        train_lbl.append(imgdata['Label'][i])
```

```python
train_img_vectors = []
for i in train_img:
    train_img_vectors.append(i.tolist())
test_img_vectors = []
for j in test_img:
    test_img_vectors.append(j.tolist())
```

```python
for i in range(360):
    for j in range(10304):
        train_img_vectors[i][j]=np.asscalar(np.asarray(train_img_vectors[i][j]))
for i in range(40):
    for j in range(10304):
        test_img_vectors[i][j]=np.asscalar(np.asarray(test_img_vectors[i][j]))
```

```python
imagesMatrix = np.matrix(train_img_vectors)
```

## Fitting PCA on train_img data such that 95% of the variance is retained

```python
from sklearn import decomposition
pca = decomposition.PCA(.95)
pca.fit(imagesMatrix)
```

```python
pca.n_components_
```

<code>Out of 10304 dimensions for an image, 176 dimensions cover 95% of variance, so we can project all the other dimensions into 176D space</code>

```python
train_img = pca.transform(np.asmatrix(train_img_vectors,dtype='float64'))
test_img = pca.transform(np.asmatrix(test_img_vectors,dtype='float64'))
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression(solver = 'lbfgs',multi_class='multinomial')
logisticRegr.fit(train_img, train_lbl)
```

```python
test_lbl[:]
```

```python
logisticRegr.predict(test_img[:])
```

### Finding Accuracy for checking model performance



```python
logisticRegr.score(test_img[:], test_lbl)
```

### Building Model - followed by Cross Validation for  Hyper Tuning

<p> Cross-validation is a technique for evaluating ML models by training several ML models on subsets of the available input data and evaluating them on the complementary subset of the data. ... In k-fold cross-validation, you split the input data into k subsets of data (also known as folds). In sklearn, we have different Kfold CV techniques available.<br>
<code><b>StratifiedKFold</b>
The splitting of data into folds may be governed by criteria such as ensuring that each fold has the same proportion of observations with a given categorical value, such as the class outcome value. This is called stratified cross-validation
<b>GroupKFold</b>
K-fold iterator variant with non-overlapping groups.
<b>RepeatedKFold</b>
Repeats K-Fold n times</code>

#### Building Model - one-versus-all classifier

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, StratifiedKFold,KFold
```

```python
from sklearn.model_selection import cross_validate
scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score,average='micro'),
           'recall' : make_scorer(recall_score,average='micro'), 
           'f1_score' : make_scorer(f1_score,average='micro')}


img_vectors = np.matrix(imgdata['Image']) 

lbl = np.matrix(imgdata['Label']) 


model = LogisticRegression(random_state=0,solver='lbfgs',multi_class='ovr').fit(train_img,train_lbl)

```

```python
skf=StratifiedKFold(n_splits=9, random_state=0)
results = cross_validate(estimator=model,
                                          X=train_img,
                                          y=train_lbl,
                                          cv=skf,
                                          scoring=scoring)
CVResultsdf=pd.DataFrame.from_dict(results)
CVResultsdf
```

### average per-class F1 score, recall and precision of the trained classifier

```python
Avg_Precision = CVResultsdf.test_precision.mean()
Avg_Recall = CVResultsdf.test_recall.mean()
Avg_F1Score = CVResultsdf.test_f1_score.mean()
print ("Avg_Precision = {} ,Avg_Recall = {} ,Avg_F1Score = {}".format(Avg_Precision,Avg_Recall,Avg_F1Score))
```

#### Method2 : Using cross_val_score, we can only get one metric, here, I am using F1_Score

```python
from sklearn.model_selection import  cross_val_score
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score


model = LogisticRegression(random_state=0,solver='lbfgs',multi_class='ovr').fit(train_img,train_lbl)
models = [model]
names = ["Logistic Regression"]

def getScores(estimator, x, y):
    yPred = estimator.predict(x)
    return (accuracy_score(y, yPred), 
            precision_score(y, yPred, average='micro'), 
            recall_score(y, yPred, average='micro'),
           f1_score(y, yPred, average='micro'))

def my_scorer(estimator, x, y):
    a, p, r ,f1 = getScores(estimator, x, y)
    return f1

scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score)}    
    
for model, name in zip(models, names):
    print (name)
    start = time.time()
    cv_score = cross_val_score(model, train_img, train_lbl,scoring=my_scorer, cv=9)
    print(cv_score)
    Avg_F1Score = cv_score.mean()
    print ('\nAvg_F1Score:',Avg_F1Score, '\n\n')
```
