## Data Preprocessing

### Imputation : NULL values are replaced with mean


```python
updated_data=df.fillna(df.mean())
```

```python
updated_data.describe()
```

## Removal of Unique Value Column


```python
unq = df.nunique()==1
```

```python
unq.head(10)
```

```python
reduced_data=updated_data.drop(list(df.columns[unq]),axis=1)
```

```python
reduced_data.shape
```

# Principle Component Analysis (PCA)

## Step1 : Standardization

```python
from sklearn.preprocessing import StandardScaler
norm_data=StandardScaler().fit_transform(reduced_data)
norm_data.shape
```

## Step2 : Finding Covariance Matrix

```python
matr = np.matmul(norm_data.T,norm_data)
```

```python
matr.shape
```

## Step3 : Finding Eigen Values and Eigen Vectors

```python
from scipy.linalg import eigh
```

```python
vals,vects = eigh(matr)
```

```python
vects.shape
```

```python
vals
```

## PCA Using sklearn

```python
from sklearn import decomposition
pca = decomposition.PCA(norm_data)
```

```python
pca.n_components = 474
pca_data = pca.fit_transform(norm_data)
```

```python
pca.explained_variance_ 
```

### Amount of variance that is preserved across each principal component

```python
each_pc_explained_variance_ratio=pca.explained_variance_ratio_
each_pc_explained_variance_ratio
```

```python
cumulativeVariance = pca.explained_variance_ratio_.cumsum()
```


### So from graph we can choose any number between 200 to 220 in order to get 97% to 99% of variance


```python
plt.figure(1, figsize=(6, 4))
plt.clf()
plt.plot(cumulativeVariance, linewidth=2)
plt.axis('tight')
plt.grid(True)
plt.xlabel('n_components')
plt.ylabel('cumulativeVariance')
plt.show()
```

```python
pca.n_components=7
principal7Components = pca.fit_transform(norm_data)
principalDF = pd.DataFrame(data=principal7Components,columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7'])
```
